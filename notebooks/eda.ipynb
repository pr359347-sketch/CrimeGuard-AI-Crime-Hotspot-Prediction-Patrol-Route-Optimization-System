# CrimeGuard-AI EDA: Environmental & Socioeconomic Features
Analysis of weather, event, and socioeconomic data distributions.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dummy data for demonstration
weather_df = pd.DataFrame({'temp': [20, 22, 25, 18, 19], 'precipitation': [0, 0.1, 0, 0.5, 0.2]})
socio_df = pd.DataFrame({'income': [45000, 55000, 60000, 35000, 48000]})

# Visualize weather distributions
plt.figure(figsize=(10, 4))
sns.histplot(weather_df['temp'], kde=True, label='Temperature')
plt.title('Distribution of Weather Features')
plt.legend()
plt.show()

# Visualize socioeconomic distributions
plt.figure(figsize=(10, 4))
sns.boxplot(x=socio_df['income'])
plt.title('Distribution of Socioeconomic Features (Income)')
plt.show()
```
